```python
def normalize_address(address):
    if not isinstance(address, str):
        return ""
    first_pass = (
        address.upper()
               .replace("POST OFFICE", "PO")
               .replace("P.O.", "PO")
               .replace(".", " ")
               .replace(",", " ")
               .replace("-", " ")
    )
    return " ".join(
        address_replacement.get(word, word)
        for word in first_pass.split()
        if word != "C/O"
    )
```

```python
cik["normalized_address"] = cik["street"].apply(normalize_address).str[:] + " " + cik["secondary"].apply(normalize_address).str[:]
cik["normalized_city"] = cik["city"].apply(lambda x: x.upper() if isinstance(x, str) else "").apply(lambda x: cik_city.get(x, x))
cik["normalized_state_or_country"] = cik["state_or_country"].apply(lambda x: x.upper() if isinstance(x, str) else "").apply(lambda x: cik_country.get(x, x))
```

```python
lei["normalized_legal_address"] = np.where(
    lei["Entity.LegalAddress.AdditionalAddressLine.1"].apply(lambda x: isinstance(x, str) and x != ""),
    lei["Entity.LegalAddress.AdditionalAddressLine.1"].apply(normalize_address) + " " +
    lei["Entity.LegalAddress.AdditionalAddressLine.2"].apply(normalize_address) + " " +
    lei["Entity.LegalAddress.AdditionalAddressLine.3"].apply(normalize_address),
    lei["Entity.LegalAddress.FirstAddressLine"].apply(normalize_address),
)
lei["normalized_hq_address"] = np.where(
    lei["Entity.HeadquartersAddress.AdditionalAddressLine.1"].apply(lambda x: isinstance(x, str) and x != ""),
    lei["Entity.HeadquartersAddress.AdditionalAddressLine.1"].apply(normalize_address) + " " +
    lei["Entity.HeadquartersAddress.AdditionalAddressLine.2"].apply(normalize_address) + " " +
    lei["Entity.HeadquartersAddress.AdditionalAddressLine.3"].apply(normalize_address),
    lei["Entity.HeadquartersAddress.FirstAddressLine"].apply(normalize_address),
)
lei["normalized_legal_city"] = lei["Entity.LegalAddress.City"].str.upper()
lei["normalized_hq_city"] = lei["Entity.HeadquartersAddress.City"].str.upper()
lei["normalized_legal_state_or_country"] = np.where(
    lei["Entity.LegalAddress.Country"].isin(["US", "CA"]),
    lei["Entity.LegalAddress.Region"].apply(lambda x: ", ".join(x.split("-")[::-1]) if isinstance(x, str) else ""),
    lei["Entity.LegalAddress.Country"],
)
lei["normalized_hq_state_or_country"] = np.where(
    lei["Entity.HeadquartersAddress.Country"].isin(["US", "CA"]),
    lei["Entity.HeadquartersAddress.Region"].apply(lambda x: ", ".join(x.split("-")[::-1]) if isinstance(x, str) else ""),
    lei["Entity.HeadquartersAddress.Country"],
)
```

```python
lei_hq = lei.query("""
normalized_legal_address != normalized_hq_address or
normalized_legal_city != normalized_hq_city or
normalized_legal_state_or_country != normalized_hq_state_or_country
""".replace("\n", " "))
```

```python
cik_lei_legal = cik.merge(
    lei,
    left_on=["normalized_address", "normalized_city", "normalized_state_or_country"],
    right_on=["normalized_legal_address", "normalized_legal_city", "normalized_legal_state_or_country"],
    how="inner",
)
cik_lei_legal["matched_legal"] = True
cik_lei_hq = cik.merge(
    lei_hq,
    left_on=["normalized_address", "normalized_city", "normalized_state_or_country"],
    right_on=["normalized_hq_address", "normalized_hq_city", "normalized_hq_state_or_country"],
    how="inner",
)
cik_lei_hq["matched_legal"] = False

cik_lei = pd.concat([cik_lei_legal, cik_lei_hq])
```
